```python
from math import gcd
import random
n = 100 
range_inc =  range(1, n + 1)
answers = [x  for x in range(1, n + 1) if gcd(x, 100) == 1]

# Check with some numbers the theorem:
# any two numbers that are part of the set S (answers list)
# the product of them is also coprime with n

len_answers = len(answers) - 1 
i1 = random.randint(1, len_answers)
i2 = random.randint(1, len_answers)

product_coprime = \
    answers[i1] * \
    answers[i2]

gcd(product_coprime, n) == 1
print(f"{answers[i1] = }")
print(f"{answers[i2] = }")
print(f"{product_coprime = }")
print(f"gcd({product_coprime}, {n}) = {gcd(product_coprime, n) }")
```

$$  \prod x_{ij} \equiv \prod a x_{i} (\mod n) $$

```python
# let's multiply all values from "answers"
# and compare with the product of "answers_T"
# and with "answers_T_p"

n = 20 
a = 3 
range_inc =  range(1, n + 1)
answers = [x  for x in range(1, n + 1) if gcd(x, 100) == 1]
answers_T = [a*x for x in answers]
answers_T_p = [x % n for x in answers_T]
print(f"{answers = }")
print(f"{answers_T = }")
print(f"{answers_T_p = }")
```

let $ \rho  $ be a prime number

let $\varphi(n)$ be Euler's totient function, which tells the number of elements in the set $S$

We know that 
$$ \varphi(\rho) = \rho - 1$$

Why this is true?

How is this possible?

```python
def totient(n: int) -> int:
    answers = [x  for x in range(1, n + 1) if gcd(x, n) == 1]
    return len(answers)
```
