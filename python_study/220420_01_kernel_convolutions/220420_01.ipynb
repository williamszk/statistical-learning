```python
image = np.array([
    [50,50,100,100],
    [50,50,100,100],
    [50,50,100,100],
    [50,50,100,100],
    [50,50,100,100],
])
```

```python
conv_kernel_X_axis_3x3 = np.array([
    [-1,0,1],
    [-2,0,2],
    [-1,0,1]
])

plt.imshow(conv_kernel_X_axis_3x3)
```

```python
# Meaning of convolutional
# A form or shape that is folded in curved or tortuous windings
# One of the irregular ridges on the surface of the brain and especially of the cerebrum of higher mammals
# A complication or intricacy of form, design, or structure
```

```python
conv_kernel_shape = conv_kernel_X_axis_3x3.shape
print(conv_kernel_shape)
```

```python
# I think that the kernel dimension needs to be a odd
# value, so that we know in which point the value will
# be placed

# suppose that we'll work only with 3x3 kernels
# i and j start from 1
```

```python
def apply_conv_kernel(image, conv_kernel_X_axis_3x3):
    
    conv_kernel_shape = conv_kernel_X_axis_3x3.shape
    
    x_size = conv_kernel_shape[0]
    y_size = conv_kernel_shape[1]

    image_x_dim = image.shape[0]
    image_y_dim = image.shape[1]

    conv_image = np.array([[0]*(image_y_dim-2)]*(image_x_dim-2))    
    
    for i in range(1,image_x_dim-1):
        for j in range(1,image_y_dim-1):
    #         print(i, j)
    #         plt.imshow(image[i-1:i+2, j-1:j+2])
    #         plt.show()
    #         plt.show()
            conv_image[i-1,j-1] = (image[i-1:i+2, j-1:j+2] * conv_kernel_X_axis_3x3).sum()

#     plt.imshow(conv_image)
#     plt.show()    
        
    return conv_image
```

```python
plt.imshow(apply_conv_kernel(image, conv_kernel_X_axis_3x3)) 
```
